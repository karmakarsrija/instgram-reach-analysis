```python
import matplotlib.pyplot as plt

# Summed values
values = [insta_home, insta_ht, insta_exp, insta_oth]
labels = ['From Home', 'From Hashtags', 'From Explore', 'From Other']

# Create a bar chart
plt.bar(labels, values, color=['blue', 'orange', 'green', 'red'])

# Add titles and labels
plt.title('Instagram Engagement Sources')
plt.xlabel('Source')
plt.ylabel('Total Engagement')

# Show the plot
plt.show()
```

```python
plt.scatter(instadf['Impressions'], instadf['Likes'], s=instadf['Likes'], alpha=0.5)

# Fit line
m, b = np.polyfit(instadf['Impressions'], instadf['Likes'], 1)  # Linear fit
plt.plot(instadf['Impressions'], m * instadf['Impressions'] + b, color='red')

# Titles and labels
plt.title('Scatter Plot of Likes vs Impressions')
plt.xlabel('Impressions')
plt.ylabel('Likes')

# Show the plot
plt.show()
```

```python
#linear relationship between likes and impressions
```

```python
numeric_df = instadf.select_dtypes(include=['number'])
correlation = numeric_df.corr()
print(correlation)
```

```python
plt.figure(figsize=(10, 8))

# Create a pairplot
sns.pairplot(numeric_df)
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()
```

```python
from itertools import combinations
cols = numeric_df.columns
combinations = list(combinations(cols, 2))

# Set up the matplotlib figure
fig, axes = plt.subplots(len(combinations), figsize=(8, len(combinations)*3))

# Create scatter plots for each combination
for ax, (x, y) in zip(axes, combinations):
    sns.scatterplot(data=numeric_df, x=x, y=y, ax=ax)
    ax.set_title(f'Scatter Plot: {x} vs {y}')

plt.tight_layout()
plt.show()
```

```python
correlation = numeric_df.corr()

# Create a heatmap using matplotlib
fig, ax = plt.subplots(figsize=(10, 8))

# Create the heatmap
cax = ax.matshow(correlation, cmap='coolwarm')

# Add color bar
plt.colorbar(cax)

# Set the ticks and labels
ax.set_xticks(np.arange(len(correlation.columns)))
ax.set_yticks(np.arange(len(correlation.columns)))
ax.set_xticklabels(correlation.columns, rotation=45, ha='right')
ax.set_yticklabels(correlation.columns)

# Title and labels
plt.title('Correlation Heatmap', pad=20)
plt.show()
```
